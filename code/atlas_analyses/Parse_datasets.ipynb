```python
import pandas as pd

# Full paths
colon_df = pd.read_csv("/home/mmiihkin/Downloads/colon_cell_metadata.tsv", sep="\t")
kidney_df = pd.read_csv("/home/mmiihkin/Downloads/kidney_cell_metadata.tsv", sep="\t")
heart_df = pd.read_csv("/home/mmiihkin/Downloads/heart_cell_metadata.tsv", sep="\t")

print("Colon:", colon_df.shape)
print("Kidney:", kidney_df.shape)
print("Heart:", heart_df.shape)
colon_df.head()
```

```python
import matplotlib.pyplot as plt
# Combine all datasets
combined_df = pd.concat([colon_df, kidney_df, heart_df], ignore_index=True)

# Count abundance of each cell type
cell_counts = combined_df["cell_type"].value_counts()

# Sort in decreasing order (value_counts already does this, but explicit is fine)
cell_counts = cell_counts.sort_values(ascending=False)

# Plot
plt.figure()
plt.bar(range(len(cell_counts)), cell_counts.values)

plt.xlabel("Cell types (sorted by decreasing abundance)")
plt.ylabel("Abundance")
plt.title("Cell type abundance across all datasets")

plt.show()
```
