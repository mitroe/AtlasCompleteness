```python
#Blood male 1.3 M cells

import scanpy as sc
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Zuckerberg spinal cord
file_path = Path.home() / "Downloads" / "c3d1a5e6-780b-4fe9-a39b-1864f927e87b.h5ad"

# Load the AnnData object
adata = sc.read_h5ad(file_path)
```

```python
counts = adata.obs["cell_type"].value_counts().sort_values(ascending=False)
    
plt.figure(figsize=(10, 5))
plt.bar(range(len(counts)), counts.values, color="gray")
    
plt.xticks(
range(len(counts)),
    counts.index,
    rotation=90,
     fontsize=14
    )
    
plt.xlabel("Cell type (sorted by decreasing abundance)", fontsize = 16)
plt.ylabel("Abundance", fontsize = 16)
plt.title("Blood cell type abundance", fontsize = 18)

    
plt.tight_layout()
plt.savefig("blood_abundance.svg", dpi=600, format="svg")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm, kstest

counts = adata.obs["cell_type"].value_counts().sort_values(ascending=False)
y = counts.values.astype(float)
y = y[y >= 0]
y_sorted = np.sort(y)[::-1]
S = len(y_sorted)

rank = np.arange(1, S + 1)
p = (S - rank + 0.5) / S  # same as your R

# Fit lognormal to the abundance values (positive only)
y_pos = y_sorted[y_sorted > 0]
shape, loc, scale = lognorm.fit(y_pos, floc=0)

# Quantile curve (rank -> predicted abundance)
pred = lognorm.ppf(p, shape, loc=loc, scale=scale)

# KS test (on values, not ranks)
D, pval = kstest(y_pos, 'lognorm', args=(shape, loc, scale))

plt.figure(figsize=(10, 5))
plt.bar(range(S), y_sorted, color="gray", label="Observed")
plt.plot(range(S), pred, linewidth=2, label=f"Lognormal quantile fit (KS p={pval:.3g})")

plt.xticks(range(S), counts.index, rotation=90, fontsize=12)
plt.xlabel("Cell type (sorted by decreasing abundance)", fontsize=14)
plt.ylabel("Abundance", fontsize=14)
plt.title("Blood cell type abundance with lognormal quantile overlay", fontsize=16)
plt.legend()
plt.tight_layout()
plt.show()
```
